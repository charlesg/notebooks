# QSTK Tutorial 1

Sample from QSTK's tutorial 1 on the website.

The first part of the code import the required libraries.

```python
import QSTK.qstkutil.qsdateutil as du
import QSTK.qstkutil.tsutil as tsu
import QSTK.qstkutil.DataAccess as da

import datetime as dt
import matplotlib.pyplot as plt
import pandas as pd
```

This next section defines a list of symbols we a looking to build a dataset as well as a date range. We grab a closing price and store in <i>na_price</i>. Following this, we normalize the prices using NumPy:

```python
ls_symbols = ["AAPL", "GLD", "GOOG", "$SPX", "XOM"]
dt_start = dt.datetime(2012, 1, 1)
dt_end = dt.datetime(2012, 12, 31)
dt_timeofday = dt.timedelta(hours=16)
ldt_timestamps = du.getNYSEdays(dt_start, dt_end, dt_timeofday)

c_dataobj = da.DataAccess('Yahoo')
ls_keys = ['open', 'high', 'low', 'close', 'volume', 'actual_close']
ldf_data = c_dataobj.get_data(ldt_timestamps, ls_symbols, ls_keys)
d_data = dict(zip(ls_keys, ldf_data))

na_price = d_data['close'].values
na_normalized_price = na_price / na_price[0, :]
```

The first plit is a simple plot of all 5 sets of closing prices for the time queried:

```python
# Where we plot the image                                                                                                                                                                    
plt.clf()
plt.plot(ldt_timestamps, na_price)
plt.legend(ls_symbols)
plt.ylabel('Adjusted Closes')
plt.xlabel('Date')
#plt.savefig('adjustedclose.png', format='png')
```

This next plot is the same plot, but this time with the <b>normalized</b> prices:

```python
# Where we plot the image                                                                                                                                                                    
plt.clf()
plt.plot(ldt_timestamps, na_normalized_price)
plt.legend(ls_symbols)
plt.ylabel('Normalized Adjusted Close')
plt.xlabel('Date')
# plt.savefig('normadjustedclose.png', format='png')
```

The next step is to compare daily returns. For this I change the symbol list to include only <b>$SPX</b> and <b>XOM</b>. I pull closing prices again, re-normalize them and compute daily returns with returnize0():

```python
# Limited symbols for returns                                                                                                                                                                
ret_symbols = ["$SPX", "XOM"]
ret_data = c_dataobj.get_data(ldt_timestamps, ret_symbols, ls_keys)
ret_data = dict(zip(ls_keys, ret_data))

ret_na_price = ret_data['close'].values
ret_na_normalized_price = ret_na_price / ret_na_price[0, :]
ret_na_rets = ret_na_normalized_price.copy()
tsu.returnize0(ret_na_rets)

# Where we plot the image                                                                                                                                                                    
plt.clf()
plt.plot(ldt_timestamps, ret_na_rets)
plt.legend(ret_symbols)
plt.ylabel('Daily Returns')
plt.xlabel('Date')
plt.savefig('dailyrets.png', format='png')
```
